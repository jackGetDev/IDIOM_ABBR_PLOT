```python
import json
import matplotlib.pyplot as plt
import string

with open('data.json') as file:
    data = json.load(file)

unique_first_letters = sorted(set(key[0] for key in data.keys()))

grouped_data = {}
for letter in unique_first_letters:
    grouped_data[letter] = [data_entry for key, data_entry in data.items() if key.startswith(letter)]

num_subplots = len(unique_first_letters)
fig, axs = plt.subplots(num_subplots, 1, figsize=(100, 100))


for i, letter in enumerate(unique_first_letters):
    if letter in grouped_data:
        data_entries = grouped_data[letter]
        bars = axs[i].bar(range(len(data_entries)), [len(data_entry) for data_entry in data_entries])
        axs[i].bar(range(len(data_entries)), [len(data_entry) for data_entry in data_entries])
        axs[i].set_xticks(range(len(data_entries)))
        axs[i].set_xticklabels([key for key in data.keys() if key.startswith(letter)],rotation=90)
        axs[i].set_ylabel('Jumlah Frasa')
        axs[i].set_title(f'Huruf {letter}')

        for bar in bars:
            height = bar.get_height()
            axs[i].text(bar.get_x() + bar.get_width() / 2, height, str(int(height)), ha='center', va='bottom')

    else:
        axs[i].axis('off')

plt.tight_layout()
plt.show()
```
